# Visualization of the function $\large f(x_1, x_2) =  x_1^2 +2 x_2^2$

```python
# 3D Scatter plot
a= np.linspace(-10,10,50)
b=a
(c,d)=np.meshgrid(a,b)
z=Function((c,d))

fig=plt.figure(figsize=(6,6))
ax=plt.axes(projection="3d")
ax.scatter3D(c,d,z,cmap="binary")  # cmap is not working here 
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(30, 30) ## to rotate the figure 
ax.set_title("$ f(x_1, x_2) =  x_1^2 +2 x_2^2 $ ",fontdict={ "fontsize": 20 , "color": "green" , "horizontalalignment":"right"})

plt.show()
```

```python
# 3D contour
a= np.linspace(-15,15,100)
b=a
(c,d)=np.meshgrid(a,b)
z=Function((c,d))

fig=plt.figure(figsize=(8,8))
ax=plt.axes(projection="3d")
ax.contour3D(c,d,z,100,cmap="Greens")
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
# ax.view_init(30, 30) ## to rotate the figure 
plt.show()
```

```python
# 3d wire frame
b=a= np.linspace(-15,15,100)
(c,d)=np.meshgrid(a,b)
z=Function((c,d))

fig=plt.figure(figsize=(8,8))
ax=plt.axes(projection="3d")        
ax.plot_wireframe(c,d,z,color="gray")

plt.show()
```

```python
# 3d surface
b=a= np.linspace(-15,15,100)
(c,d)=np.meshgrid(a,b)
z=Function((c,d))

fig=plt.figure(figsize=(8,8))
ax=plt.axes(projection="3d")
ax.plot_surface(c,d,z,cmap="viridis")   # costom cmap is working here

plt.show()
```

```python
# 3d surface with costomed rstride and cstride
b=a= np.linspace(-15,15,100)
(c,d)=np.meshgrid(a,b)
z=Function((c,d))

fig=plt.figure(figsize=(8,8))
ax=plt.axes(projection="3d")
ax.plot_surface(c,d,z,rstride=10, cstride=5,cmap="viridis")   
ax.set_title('surface',fontdict={'fontsize': 80 ,
        'fontweight': 200,
        'color': "green",
       'verticalalignment': 'baseline',
        'horizontalalignment': "left"});
plt.show()
```

```python
# 3d surface with costomed rstride and cstride
r = np.linspace(-8, 20, 200)
theta = np.linspace(-0.9 * np.pi, 0.8 * np.pi, 40)
r, theta = np.meshgrid(r, theta)

X = r * np.sin(theta)
Y = r * np.cos(theta)
z=Function((X,Y))

fig=plt.figure(figsize=(8,8))
ax=plt.axes(projection="3d")
ax.plot_surface(X,Y,z,rstride=5, cstride=3,cmap="viridis")   
ax.set_title('slice surface',fontdict={'fontsize': 80 ,
        'fontweight': 200,
        'color': "green",
       'verticalalignment': 'baseline',
        'horizontalalignment': "left"});
ax.view_init(40, 30)

plt.show()
```

```python
def perturbed_GD(F,G,X,eta=0.1,max_iteration=1000):
    function_values_list=[Function(X)]
    x1_values_list=[X[0]]
    x2_values_list=[X[1]]
    eta_values_list=[eta]
    
    iteration=0
    while (iteration<= 20):
        if (G(X)!=[0,0]).all() :
            iteration+=1
            X-= eta*G(X) - 0.08*np.random.randn(2)
            x1_values_list.append(X[0])
            x2_values_list.append(X[1])
            function_values_list.append(Function(X))
        else:
            break
    if iteration< max_iteration:
        print("\nTotal iteration nedded is   {}   which is less than the maximum iteration({})\n".format(iteration,max_iteration))
    if iteration>max_iteration:
        print("\nMaximum iteration({}) reached. Please increase the iteration to get more accurate result\n".format(max_iteration))
    return x1_values_list , x2_values_list ,function_values_list, eta_values_list 
```

## How to set the spacing between subplots in Matplotlib in Python

```python
x = [1, 2, 3]
y = [3, 2, 5]
# build subplots
figure, axes = plt.subplots(nrows=2, ncols=2)

for row in axes:
    for col in row:
        col.plot(x, y)

figure.tight_layout()
plt.show()



x = [1, 2, 3]
y = [3, 2, 5]
# build subplots
figure, axes = plt.subplots(nrows=2, ncols=2)

for row in axes:
    for col in row:
        col.plot(x, y)
figure.tight_layout(pad=4.0)
```

```python
x=np.array([1, 2, 3, 4, 5])
  
# making subplots
fig, ax = plt.subplots(2, 2)
  
# set data with subplots and plot
ax[0, 0].plot(x, x)
ax[0, 1].plot(x, x*2)
ax[1, 0].plot(x, x*x)
ax[1, 1].plot(x, x*x*x)
  
# set the spacing between subplots
plt.subplots_adjust(left=0.1,
                    bottom=0.1, 
                    right=0.9, 
                    top=0.9, 
                    wspace=0.4, 
                    hspace=0.4)
plt.show()
```

```python
from scipy.stats import multivariate_normal

X = np.linspace(-5,5,50)

Y = np.linspace(-5,5,50)

X, Y = np.meshgrid(X,Y)

X_mean = 0; Y_mean = 0

X_var = 5; Y_var = 8

pos = np.empty(X.shape+(2,))

pos[:,:,0]=X

pos[:,:,1]=Y

rv = multivariate_normal([X_mean, Y_mean],[[X_var, 0], [0, Y_var]])

fig = plt.figure()

ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, rv.pdf(pos), cmap="plasma")

plt.show()
```
